```python
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure
```

```python
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries',
         'Apples1', 'Pears1', 'Nectarines1', 'Plums1', 'Grapes1', 'Strawberries1',
         'Apples2', 'Pears2', 'Nectarines2', 'Plums2', 'Grapes2', 'Strawberries2',
         'Apples3', 'Pears3', 'Nectarines3', 'Plums3', 'Grapes3', 'Strawberries3',
         'Apples4', 'Pears4', 'Nectarines4', 'Plums4', 'Grapes4', 'Strawberries4']
years = ['2015', '2016', '2017']
```

```python
fruits
```

```python
data = {'fruits' : fruits,
        '2015'   : [2, 1, 4, 3, 2, 4]*5,
        '2016'   : [5, 3, 3, 2, 4, 6]*5,
        '2017'   : [3, 2, 4, 4, 5, 3]*5}
```

```python
# this creates [ ("Apples", "2015"), ("Apples", "2016"), ("Apples", "2017"), ("Pears", "2015), ... ]
x = [ (fruit, year) for fruit in fruits for year in years ]
counts = sum(zip(data['2015'], data['2016'], data['2017']), ()) # like an hstack
```

```python
counts
```

```python
source = ColumnDataSource(data=dict(x=x, counts=counts))
```

```python
p = figure(x_range=FactorRange(*x), plot_height=350, title="Fruit Counts by Year",
           toolbar_location=None, tools="")

p.vbar(x='x', top='counts', width=0.2, source=source)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None
```

```python
show(p)
```

```python
show(p)
```

```python

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
years = ['2015', '2016', '2017']

data = {'fruits' : fruits,
        '2015'   : [2, 1, 4, 3, 2, 4],
        '2016'   : [5, 3, 3, 2, 4, 6],
        '2017'   : [3, 2, 4, 4, 5, 3]}

# this creates [ ("Apples", "2015"), ("Apples", "2016"), ("Apples", "2017"), ("Pears", "2015), ... ]
x = [ (fruit, year) for fruit in fruits for year in years ]
counts = sum(zip(data['2015'], data['2016'], data['2017']), ()) # like an hstack

source = ColumnDataSource(data=dict(x=x, counts=counts))

p = figure(x_range=FactorRange(*x), plot_height=350, title="Fruit Counts by Year",
           toolbar_location=None, tools="")

p.vbar(x='x', top='counts', width=0.9, source=source)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None

show(p)
```
