```python
from bokeh.layouts import row
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource
```

```python
factors = ["a", "b", "c", "d", "e", "f", "g", "h"]
x =  [50, 40, 65, 10, 25, 37, 80, 60]
```

```python
source = ColumnDataSource(data=dict(x=x, factors=factors))
```

```python
dot = figure(title="Categorical Dot Plot", tools="", toolbar_location=None,
            y_range=factors, x_range=[0,100])
```

```python
dot.segment(0,"factors", "x", "factors", line_width=25, line_color="#f7ad02", source=source)
dot.circle(x="x", y="factors", source=source, size=30, fill_color="#ffffff", line_color="#f7ad02", line_width=5)
```

```python
show(dot)
```
