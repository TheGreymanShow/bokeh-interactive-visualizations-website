```python
from os.path import dirname, join

import json
import numpy as np
import pandas as pd

from bokeh.io import curdoc
from bokeh.plotting import figure
from bokeh.models.widgets import Select
from bokeh.transform import factor_cmap
from bokeh.layouts import layout, column
from bokeh.models import ColumnDataSource, Div, HoverTool

from math import pi
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
from bokeh.models import ColumnDataSource
```

```python
meta_data = {
        'x1': {
            'United States': 157,
            'United Kingdom': 93,
            'Japan': 89,
            'China': 63,
            'Germany': 44,
        },
        'x2': {
            'United States': 50,
            'United Kingdom': 100,
            'Japan': 60,
            'China': 60,
            'Germany': 55,
        }
    }
```

```python
parties_list = ['x1', 'x2']
party2 = Select(title="Party", value="x1",
               options=parties_list)
```

```python
data = pd.Series(meta_data[party2.value]).reset_index(name='value').rename(columns={'index': 'country'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = Category20c[len(meta_data[party2.value])]
```

```python
data.head()
```

```python
source = ColumnDataSource(data=dict(country=data.country, value=data.value,
                                    angle=data.angle, color=data.color))
```

```python
p = figure(plot_height=350, title="Pie Chart", toolbar_location=None,
           tools="hover", tooltips="@country: @value", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend='country', source=source)
```

```python
def update_2():
    p2.title.text = "Educational Qualification of the members of %s" % (party2.value)
    p2.title.align = 'center'

    data = pd.Series(meta_data[party2.value]).reset_index(name='value').rename(columns={'index': 'country'})
    data['angle'] = data['value'] / data['value'].sum() * 2 * pi
    data['color'] = Category20c[len(meta_data[party2.value])]
    print(data.head())
    source.data = dict(
        country=data.country,
        value=data.value,
        angle=data.angle,
        color=data.color
    )
```

```python
sizing_mode = 'scale_width'  # 'scale_width' also looks nice with this example

# Pie Chart
controls_2 = [party2]
for control in controls_2:
    control.on_change('value', lambda attr, old, new: update_2())

inputs2 = column(*controls2, sizing_mode=sizing_mode)
```

```python
l = layout([
    [inputs2, p2]], sizing_mode=sizing_mode)

update_2()  # initial load of the data

curdoc().add_root(l)
curdoc().title = "Indian Political Parties"
```

```python
p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```
